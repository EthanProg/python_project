```python
sale_data_path = data_path + '销售数据/销售单行2017'
sale_data_cols = ['SALES_ID','SEQNO','BARCODE','GODS_NAME','AMOUNT','AMONEY','APRICE','SMALL_AMOUNT','SNALL_APRICE','SMALL_BARCODE','SALES_DATE','SALES_STYLE','SHOP_ID','CREA_DATE','UPDATE_TIME']
def sale_data_all(month):
    month_str = ''
    if month < 7:  #对命名数据补齐
        month_str = '0'+str(month)
    else:
        month_str = str(month)

    sale_month_path = open(sale_data_path + str(month_str)+'.csv',encoding='utf-8')
    sale_data = pd.read_csv(sale_month_path, header=None, names = sale_data_cols,
                            dtype={'SALES_ID':np.str,'SEQNO' : np.str ,'BARCODE' : np.str ,'SMALL_BARCODE' : np.str,'SALES_DATE' : np.str,'UPDATE_TIME':np.str })
    return sale_data

sale_data_all_load = pd.read_csv(open(sale_data_path + '01.csv',encoding='utf-8'), header=None, names = sale_data_cols,
                                dtype={'SALES_ID':np.str,'SEQNO' : np.str ,'BARCODE' : np.str ,'SMALL_BARCODE' : np.str,'SALES_DATE' : np.str,'UPDATE_TIME':np.str })

for x in range(2,3):
    sale_data_all_load = pd.concat([sale_data_all_load,sale_data_all(x)]) #加载后进行数据合并
    x += 1
    if x == 7:
        break
        
sale_data_all_load.head()
```

```python
# loc通过行标签索引行数据
# loc[1]表示索引转置的是第一行
data = [[1,2,3],[4,5,6]]
index = [0,1]
columns=['a','b','c']
df = DataFrame(data=data, index=index, columns=columns)
df
```

```python
df.loc[1]
```

```python
# loc['d']表示索引转置的是第'd'行
data = [[1,2,3],[4,5,6]]
index = ['d','e']
columns=['a','b','c']
df = DataFrame(data=data,index=index,columns=columns)
df
```

```python
df.loc['d']
```

```python
# 若要loc数据会报错
```

```python
# loc可以获取多行数据
data = [[1,2,3],[4,5,6]]
index = ['d','e']
columns = ['a','b','c']
df = DataFrame(data=data,index=index,columns=columns)
df
```

```python
df.loc['d':]
```

```python
# loc索引某行某列
df.loc['d',['b','c']]
```

```python
# loc索引某列
df.loc[:,['c']]
```

```python
# iloc通过行号获取行数据
df.iloc[1]
```

```python
# 使用行标索引会报错
# df.iloc['a']
```

```python
# 通过行号可以索引多行
df.iloc[0:]
```

```python
# iloc索引列数据
df.iloc[:,[1]]
```

```python
# ix 结合前两种的混合索引
df.ix[1]
```

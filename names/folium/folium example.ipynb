```python
import folium
import pandas as pd

# folium.initialize_notebook()
```

```python
#Standard OSM
map_osm = folium.Map(location=[45.5236, -122.6750])
map_osm
```

```python
stamen = folium.Map(location=[45.5236, -122.6750], tiles='Stamen Toner',zoom_start=13)
stamen
```

```python
mt_hood = folium.Map(location=[45.372, -121.6972], zoom_start=12,
                     tiles='Stamen Terrain')
folium.Marker([45.3288, -121.6625], popup='Mt. Hood Meadows').add_to(mt_hood)
folium.Marker([45.3311, -121.7113], popup='Timberline Lodge').add_to(mt_hood)
mt_hood
```

```python
portland = folium.Map(location=[45.5236, -122.6750], tiles='Stamen Toner',
                      zoom_start=13)
folium.Marker(location=[45.5244, -122.6699], popup='The Waterfront').add_to(portland)
folium.CircleMarker(location=[45.5215, -122.6261], radius=100,
                       popup='Laurelhurst Park',
                        color='#FF0000',
                        fill_color='#3186cc',
                        weight=1,
                        fill_opacity=0.5).add_to(portland)
portland
```

```python
waypoints = folium.Map(location=[46.8527, -121.7649], tiles='Stamen Terrain',
                       zoom_start=13)
folium.Marker(location=[46.8354, -121.7325], popup='Camp Muir').add_to(waypoints)
folium.CircleMarker(location=[46.8354, -121.7325],popup='Waypoint').add_to(waypoints)
waypoints
```

```python
polygons = folium.Map(location=[45.5236, -122.6750], zoom_start=13)
folium.RegularPolygonMarker(location=[45.5012, -122.6655], popup='Ross Island Bridge',
                        fill_color='#132b5e', radius=10).add_to(polygons)
folium.RegularPolygonMarker(location=[45.5132, -122.6708], popup='Hawthorne Bridge',
                        fill_color='#45647d',  radius=10).add_to(polygons)
folium.RegularPolygonMarker(location=[45.5275, -122.6692], popup='Steel Bridge',
                        fill_color='#769d96', radius=10).add_to(polygons)
folium.RegularPolygonMarker(location=[45.5318, -122.6745], popup='Broadway Bridge',
                        fill_color='#769d96', radius=10).add_to(polygons)
polygons
```
